DATA CLEANING

```python
# CLEAN COLUMN NAMES
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# CLEAN TEXT COLUMNS
text_cols = ['customer_name', 'product_name', 'channel', 'region', 'order_status']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
# FIX DATE COLUMN
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
# CLEAN NUMERIC COLUMNS (REMOVE $, N/A)
num_cols = ['unit_price', 'shipping_cost']

for col in num_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .replace('N/A', np.nan)
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# FIX QUANTITY (REMOVE INVALID VALUES)
df = df[df['quantity'] > 0]
```

```python
# FIX BOOLEAN COLUMN 
df['is_returned'] = df['is_returned'].astype(str).str.lower().map({
    'yes': 1, 'true': 1, '1': 1,
    'no': 0, 'false': 0, '0': 0
})
```

```python
# FILLING MISSING VALUES
df['customer_segment'].fillna('unknown', inplace=True)
df['shipping_method'].fillna('unknown', inplace=True)
df['payment_method'].fillna('unknown', inplace=True)
```

```python
# FIXING DISCOUNT COLUMN
df['discount_pct'] = df['discount_pct'].fillna(0)
```

```python
# FILTERING CUSTOMER RATING
df = df[(df['customer_rating'] >= 1) & (df['customer_rating'] <= 5)]
```

```python
# RE-CALCULATING REVENUE 
df.loc[df['discount_pct'] > 1, 'discount_pct'] = df['discount_pct'] / 100
df['revenue_clean'] = df['unit_price'] * df['quantity'] * (1 - df['discount_pct'])
```

```python
#RE-CALCULATING THE PROFIT
df['profit_clean'] = (
    df['revenue_clean']
    - (df['cost_price'] * df['quantity'])
    - df['shipping_cost']
)
```

```python
df.to_csv('cleaned_ecommerce_data.csv', index=False)
```
